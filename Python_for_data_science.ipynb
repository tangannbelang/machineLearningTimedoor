```python
grocery_list = ['Bread','Cereal','Peanut Butter']
print(grocery_list)
```

```python
print(grocery_list[0]) #Menampilkan data pertama
print(grocery_list[1]) #Menampilkan data kedua
print(grocery_list[-1]) #Menampilkan data terakhir
```

```python
grocery_list.append('Sausage') #Menambahkan item pada list
grocery_list.append('Croissant')
grocery_list.remove('Bread') #Menghapus item pada list
print(grocery_list)
```

```python
grocery_list[1] = 'Strawberry Jam' #Mengubah item indeks ke
print(grocery_list)
print(len(grocery_list)) #Menampilkan jumlah item pada list
```

```python
numbers = [130,100,140,150, 190, 200, 250,200,230,250] #Do It Yourself! Coba kamu isi dengan berat anggur yang dicatat oleh Jio.
print(numbers)
```

```python
numbers.sort() #Mengurutkan item pada list menjadi urutan kecil ke besar.
print(numbers)
```

```python
grade_3 = numbers[:4] #Mengambil item sebelum index ke 4
print(grade_3)
grade_2 = numbers[4:7] #Mengurutkan item pada list menjadi urutan kecil ke besar.
print(grade_2)
grade_1 = numbers[7:] #Mengambil item setelah index ke 7
print(grade_1)
```
