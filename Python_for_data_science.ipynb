# Lesson 7 OOP


```python
class Laptop:
    def __init__(self, brand, release_year, color,RAM):
        self.laptop_brand = brand
        self.laptop_release_year = release_year
        self.laptop_color = color
        self.laptop_RAM = RAM
    def information(self):
        return f"{self.laptop_color} Laptop {self.laptop_brand} release in {self.laptop_release_year} with {self.laptop_RAM} RAM"
```

```python
#indentasi sejajar dengan class
laptop_1 = Laptop("Dell", 2021, "Grey", 8)
```

```python
#indentasi sejajar dengan class
laptop_2 = Laptop("Macbook", 2022, "Silver", 8)
```

```python
print(laptop_1.laptop_brand)
# Tambahkan code di bawah ini
print(laptop_1.information())
```

```python
grades = [100,98,97,93,91,90,88,85,80,78]
```

```python
def bubble_sort(arr):
   n = len(arr)
   for i in range(n):
       for j in range(0, n-i-1):
           if arr[j] < arr[j+1]:
               arr[j], arr[j+1] = arr[j+1], arr[j]
       print(arr)
   return arr
```

```python
sorted_grades = bubble_sort(grades)
print(sorted_grades)
```

```python
def insertion_sort(arr):
   n = len(arr)
   for i in range(1, n):
       key = arr[i]
       j = i-1
       while j >= 0 and key > arr[j]:
           arr[j+1] = arr[j]
           j -= 1
       arr[j+1] = key
       print(arr)
   return arr
```

```python
insertion_sort(grades)
```

```python
def linear_search(arr, target):
   for i in range(len(arr)):
       if arr[i] == target:
           return i
   return "Target not found in the list"
```

```python
grades.sort(reverse=True) # sorting the list in descending order
print(grades)
# call the function
index = linear_search(grades, 97)
print("Target number found at index: ", index)
```
