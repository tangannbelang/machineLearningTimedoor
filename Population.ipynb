Load dan cek data

```python
import pandas as pd

df = pd.read_csv('population.csv')
df.head()
```

cek negara dengan populasi terbanyak di 2020

```python
highest = df['2020'].nlargest(20).tolist() #Mencari kolom 2020 yang memiliki nilai terbesar (di filter 20 terbesar)
filtered_highest = df[df['2020'].isin(highest)] #Melakukan filterasi dataframes yang mengandung nilai data populasi terbesar di tahun 2020
filtered_highest
```

Analisa data

```python
import seaborn as sns
sns.barplot(x='Country Name', y='2020', data=df.tail()) #Ambil data 5 terakhir untuk sample data
```

Data preparation

```python
world=df.loc[df['Country Name']=='World'] #Filter data hanya untuk Country Name = â€˜Worldâ€™
world.drop(['Country Name','Country Code', 'Indicator Name','Indicator Code'],axis=1,inplace=True) #Menghapus kolom yang tidak dipakai
world = world.T #Melakukan proses Transform dataframes
world.head()
```

```python
world.dropna(inplace=True)
world=world.reset_index().rename(columns={259:'population','index':'year'}) #Mengubah â€˜header columnâ€™ 259 menjadi population dan indexnya akan diubah menjadi year
world.tail()
```

Plot the data

```python
sns.lineplot(x='year', y='population', data=world.tail(10))
sns.scatterplot(x='year', y='population', data=world.tail(10))
```

Linear regression model

```python
world['population'] = pd.to_numeric(world['population'])
world['year'] = pd.to_numeric(world['year'])
x = world['year']
y = world['population']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
X_train = X_train.values.reshape(-1,1) #Convert menjadi 2D array
X_test = X_test.values.reshape(-1,1) #Convert menjadi 2D array
y_train = y_train.values.reshape(-1,1) #Convert menjadi 2D array
y_test = y_test.values.reshape(-1,1) #Convert menjadi 2D array
```

Fitting linear regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model = model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("True vs Predicted ")
sns.lineplot(x=X_test.reshape(-1), y=y_pred.reshape(-1))
sns.scatterplot(x=X_test.reshape(-1), y=y_pred.reshape(-1), marker="x", color="green", s=200) #Untuk nilai prediksi disimbolkan dengan Cross(x) hijau
sns.scatterplot(x=X_test.reshape(-1), y=y_test.reshape(-1), color="red") # Untuk nilai sesungguhnya disimbolkan dengan scatterplot merah
```

Predict future population

```python
test_data = model.predict([[2040]]) #Input yang dimasukannya harus dalam bentuk 2D array
test_data
```
